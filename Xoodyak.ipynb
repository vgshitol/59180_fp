```python
class Xoodyak:
  def __init__(self, msg='', msg_len=''):
    self.msg = msg
    self.msg_len = int(msg_len)
    self.hash_len = 32
    self.r=16
    self.C_D=0x03
    self.state_size = 48
    self.state=np.zeros(self.state_size,dtype=np.uint8)
    self.c = np.array([[0x12,0x00,0x00,0x00],
                       [0xa0,0x01,0x00,0x00],
                       [0xf0,0x00,0x00,0x00],
                       [0x80,0x03,0x00,0x00],
                       [0x2c,0x00,0x00,0x00],
                       [0x60,0x00,0x00,0x00],
                       [0x14,0x00,0x00,0x00],
                       [0x20,0x01,0x00,0x00],
                       [0xd0,0x00,0x00,0x00],
                       [0xc0,0x03,0x00,0x00],
                       [0x38,0x00,0x00,0x00],
                       [0x58,0x00,0x00,0x00]],dtype=np.uint8)
    self.out256 = []
    
  def convert_str_arr(self,):
    msg_lst = []
    #for i in range(0,len(self.msg),2):
    for i in range(self.msg_len):
      msg_lst.append(int(self.msg[2*i:2*i+2],16))
    self.msg_arr = msg_lst    
    
  def print_val(self,):
    print('MSG : ',self.msg)
    print('MSG_LEN : ',self.msg_len)
    print('R : ',self.r)
    print('C_D : ',self.C_D)
 #   print('MSG_ARR : ',self.msg_arr)
    print('State:',self.state)
    return 0;
   
  def down(self,x=[],c_d=0x03):
    for i in range(len(x)):
  #      print(i,x[i])
        self.state[i] ^= x[i]
        
    self.state[len(x)] ^= 0x01
    self.state[self.state_size-1] ^= (c_d & 0x01) 
    return x
    
  def up(self,extract=False,n = 1):
    state_384 = state_convert(self.state)
    state_384 = call_xoodoo(state_384)
    self.state = state_r_convert(state_384)
    
    out256 = []
    if(extract is True): 
        out256 = self.state[:16]
    return out256   
    
  def xoodoo(self,):
    for i in range(12):
        a0 = self.state[0:16]
        a1 = self.state[16:32]
        a2 = self.state[32:48]

        #theta
        p = np.zeros(16,dtype=np.uint8)
        p =  a0 ^ a1 ^ a2
    #    self.display_plane(p,label='P: ')
        e = self.cyclic_shift(p,1,5) ^ self.cyclic_shift(p,1,14)
    #    self.display_plane(e,label='E: ')
        a0 = a0 ^ e
        a1 = a1 ^ e
        a2 = a2 ^ e
     #   print(a0,a1,a2)

        # rho west
        a1 = self.cyclic_shift(a1,1,0)
        a2 = self.cyclic_shift(a2,0,11)

        #ita
        a0 = a0 ^ np.append(self.c[11-i] ,np.zeros(12,dtype=self.c.dtype))

        #Xeta
        b0 = np.bitwise_and(np.invert(a1),a2)
        b1 = np.bitwise_and(np.invert(a2),a0)
        b2 = np.bitwise_and(np.invert(a0),a1)
        a0 = a0 ^ b0
        a1 = a1 ^ b1
        a2 = a2 ^ b2

        # rho east
        a1 = self.cyclic_shift(a1,0,1)
        a2 = self.cyclic_shift(a2,2,8)
    
        self.state[0:16] = a0
        self.state[16:32] = a1
        self.state[32:48] = a2
   # self.display_state()   
    return self.state

  def cyclic_shift(self,p,t=1,v=5):
    a = []
    for x in range(4):
      z = ''
      for zb in range(4):
        z = (format(p[4*x + zb],'08b')) + z
      a.append(z)
  #  print('A: Before' ,a)
    for x in range(4):
      a[x] = (a[x][v:] + a[x][0:v])  
  #  print('A: Z After' ,a)
    a = a[-t:] + a[:-t]
  #  print('A: Z X after' ,a)
    o = []
    for i in range(4):
      for j in range(3,-1,-1):
        o.append(int(a[i][8*j:8*j+8],2))
   # print(o)
  #  self.display_plane(o,'Shift :' + str(t) + str(v))
    return np.array(o,dtype=np.uint8)


  def absorb(self,):
    # Control Signals 
    b_len = self.msg_len if (self.msg_len < self.r) else self.r
    num_offset = math.ceil(self.msg_len/16)
    offset_lst = [16*x for x in range(1,num_offset,1)]
    
    offset = 0
    x = self.msg_arr[offset:offset+b_len]
    
    print(self.msg_len,self.r,b_len)
    print('OFFSET_LST :', offset_lst,'\nX : ',x)
 
    # down operation
    d = self.down(x)
    print(self.state)
    
    for offset in offset_lst:
        b_len = self.msg_len-offset if (self.msg_len-offset < self.r) else self.r
        x = self.msg_arr[offset:offset+b_len]
        print('X :',x,'B : ',b_len)
        u = self.up()
        d = self.down(x,0x00) 
  #      print('State : ',self.state)
    return self.state
 
  def display_state(self,):
    for k in range(3):
      for j in range(4):
        print('0x',end='')
        for i in range(4):
          print("{0:#0{1}x}".format((self.state[4*4*k + 4*j + i]),4)[2:],end='')
        print()
      print() 
    
  def display_plane(self,p=[],label='Plane : '):
        print(label,end=' ')
        for j in range(4):
            print('0x',end='')
            for i in range(4):
              print("{0:#0{1}x}".format((p[4*j + i]),4)[2:],end='')
            print('\t',end='')
        print()    
    
  def display_output(self,o=[],l=32):
        if (len(o)==0):
            o = self.output
            
        print('0x',end='')
        for i in range(l):
          print("{0:#0{1}x}".format((o[i]),4)[2:],end='')
        print()
        
  def squeeze(self,):
    # Control Signals
    b_len = 16
    o1 = self.up(True)
    o1 = np.array(o1)
    self.down(c_d=0x00)
    o2 = self.up(True)
    o2 = np.array(o2)
    self.output = np.append(o1,o2)
    return self.output
        
```
